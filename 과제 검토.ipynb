```python
class account:
    def __init__(self,name,money):
        self.name=name
        self.balance=int(money)
        print('Great!!')
    def deposit(self,money):
        money=int(money)
        self.balance+=money
    def withDraw(self,money):
        money=int(money)
        if self.balance>=money:
            self.balance-=money
        else:
            print("잔고가 부족하여",commanum(self.balance),"만큼만 인출되었습니다.")
            self.balance=0
    def showBalance(self):
        return self.balance
    def showName(self):
        return self.name

```

```python
def commanum(X):
    X=str(X)
    if len(X)>=3:    
        y=list(X)
        n=[',']+y[-3:]
        for i in range(2,len(y)//3+1):
            x=[',']+y[-3*i:-3*i+3]
            n=x+n
        n=y[:len(X)%3]+n
        if n[0]==',':
            n=n[1:]
        nu=''.join(n)
        return nu    
    elif len(X)<3:
        return X
```
